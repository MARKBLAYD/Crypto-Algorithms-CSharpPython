```python
# Побитовый xor

# Выполняет побитовую операцию XOR (исключающее ИЛИ) между двумя массивами данных.
def x_oper(ar1, ar2):
    # Аргументы:
    # ar1, ar2 - два массива битов одинаковой длины.
    # Возвращает результат побитовой операции XOR (массив битов).

    data = [0] * len(ar1)  # Инициализируем пустой список для хранения результата.

    # Проходим по каждому биту в массивах и выполняем операцию XOR.
    for i in range(0, len(ar1)):
        if ar1[i] == ar2[i]:  # Если биты одинаковые, результат 0.
            data[i] = 0
        else:  # Если биты разные, результат 1.
            data[i] = 1

    return data  # Возвращаем результат операции XOR.
```

```python
def g_oper(n, h, m):
    # Реализует криптографическое преобразование, основанное на алгоритме ГОСТ Стрибог.
    # Аргументы:
    # n - начальное значение (массив битов, длина 512 бит).
    # h - текущее значение хеша (массив битов, длина 512 бит).
    # m - блок сообщения (массив битов, длина 512 бит).
    # Возвращает результат выполнения преобразования (массив битов длиной 512).

    # Генерация ключа K на основе текущего значения хеша и входного параметра n.
    k = l_oper(p_zamn(s_zamn(x_oper(h, n))))  

    st = m  # Инициализация переменной st как текущего блока сообщения.

    for i in range(0, 12):  # Выполняем 12 раундов преобразования.
        st = l_oper(p_zamn(s_zamn(x_oper(st, k))))  # Применяем раундовые функции (подстановка, перестановка, линейное преобразование).
        k = vch_key(k, i)  # Генерация нового ключа для текущего раунда.

    # После 12 раундов вычисляем XOR между st и последним ключом.
    st = x_oper(st, k)

    # Возвращаем результат выполнения XOR с текущим значением хеша и блоком сообщения.
    return x_oper(x_oper(st, h), m)
```

```python
# Преобразует массив битов (0 и 1) в десятичное число (int).
def mas_to_int(byte_array):
    # Аргумент:
    # byte_array - список битов (0 и 1), представляющий двоичное число.
    # Возвращает десятичное представление этого числа.

    p = 0  # Инициализируем результат.

    for i in range(0, len(byte_array)):
        # Постепенно формируем число, сдвигая результат влево и добавляя текущий бит.
        p = byte_array[i] + 2 * p

    return p  # Возвращаем результат в десятичном формате.
```

```python
# Преобразует десятичное число (int) в массив битов (0 и 1) для представления числа длиной 64 байта (512 бит).
def int_to_mas(value,raz):
    ter = [0] * raz  # Инициализируем массив длиной 512 бит (все элементы равны 0).
    d = raz-1  # Индекс для записи битов, начиная с младшего разряда (правого конца массива).

    while value > 0:  # Выполняем, пока число не станет равно 0.
        if value % 2 == 1:  # Проверяем младший бит числа.
            ter[d] = 1  # Устанавливаем соответствующий бит в массиве в 1.
        value //= 2  # Убираем младший бит, сдвигая число вправо (деление на 2).
        d -= 1  # Переходим к следующей позиции в массиве (слева от текущей).

    return ter  # Возвращаем массив битов.
```

```python
def oper_512(a, b):
    # Функция выполняет сложение двух чисел с ограничением по модулю 2^512.
    # Аргументы:
    # a - первое число (целое).
    # b - второе число (целое).
    # Возвращает массив байтов длиной 64, представляющий результат сложения.

    # Складываем числа и берем остаток от деления на 2^512 (операция по модулю).
    result = (a + b) % (2 ** 512)

    # Преобразуем результат в массив байтов длиной 64 и возвращаем.
    return int_to_mas(result,512)
```

```python
def bigkey_mas(key):
    # Функция преобразует строку, представляющую ключ в шестнадцатеричном формате,
    # в массив битов (0 и 1).
    # Аргумент:
    # key - строка, содержащая шестнадцатеричные символы (0-9, a-f).
    # Возвращает список битов (длиной 4 бита на каждый символ).

    tr = [0] * len(key) * 4  # Инициализируем массив для хранения битов.

    for i in range(0, len(key)):
        # Используем оператор match для обработки каждого символа ключа.
        match key[i]:
            case '0': tr[i * 4:i * 4 + 4] = [0, 0, 0, 0]  # Символ '0' -> [0, 0, 0, 0]
            case '1': tr[i * 4:i * 4 + 4] = [0, 0, 0, 1]  # Символ '1' -> [0, 0, 0, 1]
            case '2': tr[i * 4:i * 4 + 4] = [0, 0, 1, 0]  # Символ '2' -> [0, 0, 1, 0]
            case '3': tr[i * 4:i * 4 + 4] = [0, 0, 1, 1]  # Символ '3' -> [0, 0, 1, 1]
            case '4': tr[i * 4:i * 4 + 4] = [0, 1, 0, 0]  # Символ '4' -> [0, 1, 0, 0]
            case '5': tr[i * 4:i * 4 + 4] = [0, 1, 0, 1]  # Символ '5' -> [0, 1, 0, 1]
            case '6': tr[i * 4:i * 4 + 4] = [0, 1, 1, 0]  # Символ '6' -> [0, 1, 1, 0]
            case '7': tr[i * 4:i * 4 + 4] = [0, 1, 1, 1]  # Символ '7' -> [0, 1, 1, 1]
            case '8': tr[i * 4:i * 4 + 4] = [1, 0, 0, 0]  # Символ '8' -> [1, 0, 0, 0]
            case '9': tr[i * 4:i * 4 + 4] = [1, 0, 0, 1]  # Символ '9' -> [1, 0, 0, 1]
            case 'a': tr[i * 4:i * 4 + 4] = [1, 0, 1, 0]  # Символ 'a' -> [1, 0, 1, 0]
            case 'b': tr[i * 4:i * 4 + 4] = [1, 0, 1, 1]  # Символ 'b' -> [1, 0, 1, 1]
            case 'c': tr[i * 4:i * 4 + 4] = [1, 1, 0, 0]  # Символ 'c' -> [1, 1, 0, 0]
            case 'd': tr[i * 4:i * 4 + 4] = [1, 1, 0, 1]  # Символ 'd' -> [1, 1, 0, 1]
            case 'e': tr[i * 4:i * 4 + 4] = [1, 1, 1, 0]  # Символ 'e' -> [1, 1, 1, 0]
            case 'f': tr[i * 4:i * 4 + 4] = [1, 1, 1, 1]  # Символ 'f' -> [1, 1, 1, 1]

    return tr  # Возвращаем массив битов.
```

```python
def to_bytes(data1):
    # Функция преобразует двоичный массив в строку байтов в шестнадцатеричном формате.
    # Аргумент:
    # data1 - список битов (0 и 1).
    # Возвращает строку, представляющую данные в виде последовательности шестнадцатеричных байтов.

    data = data1  # Копируем входные данные.

    byte_array = []  # Инициализируем массив для хранения байтов.
    for i in range(0, len(data), 8):
        byte = data[i:i + 8]  # Извлекаем 8 бит для формирования одного байта.
        byte_value = int("".join(map(str, byte)), 2)  # Конвертируем биты в десятичное число.
        byte_array.append("%02x" % byte_value)  # Форматируем байт в шестнадцатеричную строку и добавляем в массив.

    return ''.join(byte_array)  # Объединяем все байты в одну строку и возвращаем.
```

```python
def hesh_f(mm, bb):
    # Функция реализует хеширование данных согласно алгоритму ГОСТ Стрибог.
    # Аргументы:
    # mm - входное сообщение в двоичном представлении (список битов).
    # bb - длина хеша (256 или 512 бит).

    if bb == 512:
        h = [0] * 512  # Инициализация начального значения хеша для 512-битного режима.
    else:
        h = [0, 0, 0, 0, 0, 0, 0, 1] * 64  # Инициализация начального значения хеша для 256-битного режима.

    n = [0] * 512  # Счетчик обработанных битов (инициализируется нулями).
    e = [0] * 512  # Контрольная сумма (инициализируется нулями).

    while len(mm) >= 512:
        mt = mm[-512:]  # Выделение блока данных размером 512 бит.
        h = g_oper(n, h, mt)  # Применение нелинейного преобразования к текущему состоянию хеша.
        n = oper_512(mas_to_int(n), 512)  # Увеличение счетчика обработанных битов на 512.
        e = oper_512(mas_to_int(e), mas_to_int(mt))  # Обновление контрольной суммы.
        mm = mm[:-512]  # Удаление обработанного блока данных.

    pop = len(mm)  # Длина оставшегося сообщения (менее 512 бит).
    mm = [0] * (511 - pop) + [1] + mm  # Дополнение сообщения (добавление единичного бита и нулей).
    h = g_oper(n, h, mm)  # Применение нелинейного преобразования к последнему блоку.
    n = oper_512(mas_to_int(n), pop)  # Обновление счетчика с учетом длины последнего блока.
    e = oper_512(mas_to_int(e), mas_to_int(mm))  # Обновление контрольной суммы для последнего блока.

    # Завершающие преобразования с использованием счетчика и контрольной суммы.
    h = g_oper([0] * 512, h, n)
    h = g_oper([0] * 512, h, e)

    if bb == 512:
        return h  # Возврат 512-битного хеша.
    else:
        return h[:256]  # Возврат 256-битного хеша (первые 256 бит).
```

Лабораторная работа по написанию программы реализующей ЭЦП по ГОСТ Р 34.10─2018 (Элептические кривые, 512,1024 - бит)

```python
# Подфункция -1 степени по алгоритму Евклида (нахождение обратного элемента по модулю)
def g_ev(a, b):
    # В m сохраняем частное от деления b на a (первый шаг алгоритма Евклида)
    m = b // a  # Частное от деления b на a

    # l - это остаток от деления b на a, то есть b - m * a
    l = b - a * m  # Остаток от деления b на a

    g = 1  # Изначально g равно 1, оно будет изменяться в процессе алгоритма
    # Если остаток не равен 1, продолжаем вычисления по рекурсии
    if l != 1:
        # Рекурсивно вызываем функцию для l и a (по сути, мы решаем задачу для меньших чисел)
        k, g = g_ev(l, a)
        # Обновляем значение g, учитывая результат рекурсии
        g = k * m + g
        return g, k  # Возвращаем обновленные значения g и k (обратное число и коэффициент)

    # Если остаток равен 1, то мы нашли обратное число, возвращаем его
    else:
        return m, g  # Возвращаем m и g, где m - это коэффициент обратного числа, а g - само обратное число
```

```python
# Функция для нахождения обратного числа по алгоритму Евклида
def alg_ev(a, b):
    # Шаг 1: Используем вспомогательную функцию g_ev для нахождения чисел n и m
    n, m = g_ev(a, b)
    
    # Шаг 2: Проверка, если решение корректно (для случая, когда обратное число существует)
    if a * n - m * b != 1:
        # Если условие не выполнено (это означает, что a и b не взаимно просты), возвращаем b - n
        return b - n
    else:
        # Если условие выполнено, то n - это искомое обратное число
        return n
```

```python
# Функция сложения точек P и Q на эллиптической кривой
def sl_tchk(xp, yp, xq, yq, a, p):
    # Инициализация координат результата
    xr = None
    yr = None
    
    # Случай 1: x-координаты разных точек
    if (xp != xq):
        # Вычисление коэффициента наклона (lambda) для двух различных точек
        lam = ((yq - yp) * alg_ev((xq - xp) % p, p)) % p
        # Вычисление x-координаты точки R = P + Q
        xr = (lam * lam - xp - xq) % p
        # Вычисление y-координаты точки R = P + Q
        yr = (lam * (xp - xr) - yp) % p
    
    # Случай 2: x-координаты одинаковы, y-координаты равны (точка P == точка Q)
    elif (yp == yq):
        # Вычисление коэффициента наклона (lambda) для случая P = Q
        lam = ((3 * xp * xp + a) * alg_ev((2 * yp) % p, p)) % p
        # Вычисление x-координаты точки R = P + P
        xr = (lam * lam - xp - xq) % p
        # Вычисление y-координаты точки R = P + P
        yr = (lam * (xp - xr) - yp) % p

    # Возвращаем координаты результирующей точки R
    return xr, yr
```

```python
# Функция умножения точки на скаляр k
def ym_tchk(xp, yp, k, a, p):
    # Инициализация переменной для работы с двоичным разложением числа k
    ter = k
    mas = []  # Список для представления числа k в двоичной форме
    
    # Двоичное разложение числа k
    while ter > 0:
        if ter % 2 == 1:
            mas = [1] + mas  # Если текущий бит равен 1, добавляем 1 в начало списка
            ter = ter - 1    # Уменьшаем число на 1
        else:
            mas = [0] + mas  # Если текущий бит равен 0, добавляем 0 в начало списка
        ter = ter // 2     # Делим число на 2 для перехода к следующему биту
    
    # Инициализация начальной точки (начинаем с точки P)
    ter = 0
    x2 = xp
    y2 = yp
    # Переменные для хранения координат результирующей точки
    for i in range((len(mas) - 1), -1, -1):  # Обрабатываем биты числа k с конца
        if mas[i] == 1:  # Если текущий бит равен 1
            if ter == 0:
                # Если терминировали умножение, начинаем с первой точки
                xq = x2
                yq = y2
                ter = 1
            else:
                # Иначе складываем текущую точку с результирующей
                xq, yq = sl_tchk(xq, yq, x2, y2, a, p)
        
        # Умножение точки на 2 (по принципу "дублирования")
        x2, y2 = sl_tchk(x2, y2, x2, y2, a, p)  # Точка удваивается в каждом шаге

    return xq, yq  # Возвращаем результат
```

```python
import random

# Функция создания цифровой подписи (чисел r и s) для сообщения MES
def SCP(MES, raz, q, xp, yp, a, p, d):
    # Шаг 1. Вычисление хеша сообщения
    hesh = hesh_f(bigkey_mas(MES), (raz // 2))  # Преобразование сообщения в хеш (половина разрядности)

    # Шаг 2. Вычисление a и e
    alf = mas_to_int(hesh)  # Преобразование хеша в целое число
    e = alf % q  # Вычисление e как остатка от деления хеша на q
    if e == 0:  # Если e равно 0, меняем на 1 (для избежания проблем с делением)
        e = 1
    
    ttt = True  # Инициализация флага продолжения цикла

    while ttt:  # Шаг 3. Генерация случайного числа k и вычисление подписи
        # Шаг 3. Генерация случайного числа k
        k = random.randint(1, q - 1)  # Генерация случайного числа k, которое меньше q

        # Шаг 4. Вычисление точки эллиптической кривой C = k * P
        xc, yc = ym_tchk(xp, yp, k, a, p)  # Умножение точки P на k, получение точки C
        r = xc % q  # Вычисление r как остатка от деления x-координаты точки C на q
        if r != 0:  # Если r не равно 0, продолжаем, иначе генерируем новое k
            # Шаг 5. Вычисление s
            s = (r * d + k * e) % q  # Вычисление параметра s (с использованием закрытого ключа d)
            if s != 0:  # Если s не равно 0, выходим из цикла
                ttt = False  # Завершаем цикл

    # Шаг 6. Формирование подписи
    # Преобразуем r и s в массивы битов с разрядностью, равной половине длины подписи
    data2 = int_to_mas(r, (raz // 2))  # Преобразуем r в массив битов
    data2.extend(int_to_mas(s, (raz // 2)))  # Преобразуем s в массив битов и добавляем к r
    return data2  # Возвращаем итоговую подпись
```

```python
# Функция проверки цифровой подписи
def PCP(data2, MES, raz, q, xp, yp, xq, yq, a, p):
    
    pr = True  # Инициализация флага для успешной проверки подписи
    
    # Шаг 1. Извлечение значений r и s из подписи и проверка их корректности
    r = mas_to_int(data2[0:(raz // 2)])  # Преобразуем первые половину массива в r
    s = mas_to_int(data2[(raz // 2):raz])  # Преобразуем вторую половину массива в s
    
    # Проверка, что r и s находятся в пределах допустимых значений
    if (0 < r) and (r < q) and (0 < s) and (s < q):
        
        # Шаг 2. Вычисление хеша сообщения
        hesh = hesh_f(bigkey_mas(MES), raz // 2)  # Преобразуем сообщение в хеш (половина разрядности подписи)

        # Шаг 3. Вычисление значения e
        alf = mas_to_int(hesh)  # Преобразуем хеш в целое число
        e = alf % q  # e = хеш % q
        if e == 0:  # Если e равно нулю, заменяем на 1 (по алгоритму ГОСТ)
            e = 1

        # Шаг 4. Вычисление значения v (обратное значение e по модулю q)
        v = alg_ev(e, q)  # Находим v с помощью расширенного алгоритма Евклида

        # Шаг 5. Вычисление z1 и z2
        z1 = (s * v) % q  # z1 = s * v % q
        z2 = (-r * v) % q  # z2 = -r * v % q (обратите внимание, что знак минус используется в модульной арифметике)

        # Шаг 6. Вычисление точки C
        # Сначала находим точки, соответствующие z1 * P и z2 * Q
        xpz, ypz = ym_tchk(xp, yp, z1, a, p)  # Вычисление z1 * P
        xqz, yqz = ym_tchk(xq, yq, z2, a, p)  # Вычисление z2 * Q
        
        # Далее суммируем эти две точки
        xc, yc = sl_tchk(xpz, ypz, xqz, yqz, a, p)  # Сложение двух точек (z1 * P) и (z2 * Q)
        
        R = xc % q  # R = x-координата полученной точки (по модулю q)

        # Шаг 7. Проверка равенства R и r
        # Если R != r, то подпись неверна
        if R != r:
            return False  # Подпись не прошла проверку
        else:
            return True  # Подпись верна
    else:
        return False  # Если r или s не лежат в допустимом диапазоне, подпись неверна
```

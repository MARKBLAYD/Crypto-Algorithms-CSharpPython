```python
#Функция для читабельного вывода данных
    #На вход подается массив из 128
def to_bytes(data1):
    data=data1

    # Преобразуем массив в байты
    byte_array = []
    for i in range(0, len(data), 8):
        byte = data[i:i + 8]  # Берем 8 элементов
        byte_value = int("".join(map(str, byte)), 2)  # Конвертируем в десятичное
        byte_array.append("%02x" % byte_value)

    return ''.join(byte_array)
```

```python
# Преобразует десятичное число (int) в массив 0 и 1 для одного байта

def int_to_mas(value):
    ter = [0] * 8  # Инициализируем массив из 8 элементов
    d = 7  # Индекс для записи битов в массив (начинаем с младшего бита)

    while value > 0:  # Пока есть ненулевые биты
        if value % 2 == 1:  # Если младший бит равен 1
            ter[d] = 1
        value //= 2  # Переходим к следующему биту
        d -= 1  # Смещаемся влево

    return ter
```

```python
# Преобразует массив 0 и 1 для одного байта в десятичное число (int)

def mas_to_int(byte_array):
    # Преобразуем массив в строку, затем в десятичное число
    p=0
    for i in range(0,len(byte_array)):
        p=byte_array[i]+2*p
    return p
```

```python
# Побитовый xor

def x_oper(ar1, ar2):
    data=[0]*len(ar1)
    for i in range(0,len(ar1)):
        if ar1[i]==ar2[i]:
            data[i]=0
        else:
            data[i]=1
    return data
```

```python
# умножение двух байтов в поле Галуа GF(2^8) с полиномом x^8 + x^7 + x^6 + x + 1

def gf_mul(a1, b1): 
    a=a1
    b=b1
    c = 0  # Результат умножения
    for _ in range(8):  # Итерируемся по каждому биту
        if b & 1:  # Если младший бит b равен 1
            c ^= a  # XOR с текущим значением a
        hi_bit = a & 0x80  # Проверяем старший бит a (маска 0x80 = 10000000)
        a = (a << 1) & 0xFF  # Сдвиг a влево на 1 бит и усечение до 8 бит
        if hi_bit:  # Если старший бит был установлен
            a ^= 0xC3  # XOR с неприводимым полиномом x^8 + x^7 + x^6 + x + 1
        b >>= 1  # Сдвигаем b вправо на 1 бит
    return c
```

```python
# функция R сдвигает данные и реализует уравнение, представленное для расчета L-функции
def r_oper(data1):
    """
    Выполняет сдвиг данных на один байт влево и заполняет последний байт результатом
    суммы произведений каждого байта с числом из дополнительного списка.
    :param data: Список из 16 байтов (чисел от 0 до 255)
    :param constants: Список из 16 чисел (константы для умножения)
    :return: Новый список из 16 байтов после преобразования
    """
    data=data1
    const = [1, 148, 32, 133, 16, 194, 192, 1, 251, 1, 192, 194, 16, 133, 32, 148]  # Константы
    pp=[0]*8
    for i in range(0,16):
        yt=mas_to_int(data[i*8:i*8+8])
        pp=x_oper(pp,int_to_mas(gf_mul(yt,const[15-i])))
    for i in range(0,8): #сдвиг влево на 1 байт
        a=data[127]
        for j in range(126,-1,-1):
            data[j+1]=data[j]
        data[0]=a
    for i in range(0,8):
        data[i]=pp[i]
    return data

def obr_r(data1):
    data=data1
    # Константы из ГОСТ Р 34.12-2015
    const = [1, 148, 32, 133, 16, 194, 192, 1, 251, 1, 192, 194, 16, 133, 32, 148]
    # Последние 8 бит (a_0)
    a_0 = data[:8]
    for i in range(1, 16):
        yt = mas_to_int(data[i * 8:i * 8+8])  # Конвертация в число
        a_0 = x_oper(a_0, int_to_mas(gf_mul(yt, const[i])))
    data[:8] = a_0
    for i in range(0,8): #сдвиг влево на 1 байт
        g=data[0]
        for j in range(0,127):
            data[j]=data[j+1]
        data[127]=g
    return data
```

```python
# реализыция функции линейного приобразования

def l_oper(data1):
    data=data1
    for i in range(0,16):
        data=r_oper(data)
    return data

def obr_l(data1):
    data=data1
    for i in range(0,16):
        data=obr_r(data)
    return data
```

```python
#функция расчета констант

const_c = []
def rscht_c():
    for i in range(0,32):
        data=[0]*128
        h=int_to_mas(i+1)
        for j in range(0,8):
            data[j+120]=h[j]
        global const_c
        const_c.append(l_oper(data))
```

```python
# функция преобразования ячейки фейстиля

def feist(key1,key2,vb):
    key=[0]*128
    for i in range(0,8):
        key=x_oper(key1,const_c[vb*8+i])
        key=s_zamn(key)
        key=l_oper(key)
        key=x_oper(key2,key)
        key2=key1
        key1=key
    return key1, key2
```

```python
# вычисление раундовых ключей

round_key = []
def vch(key):
    round_key.append(key[:128])
    round_key.append(key[128:])
    key1=key[:128]
    key2=key[128:]
    for i in range(0,4):
        key1,key2=feist(key1,key2,i)
        round_key.append(key1)
        round_key.append(key2)
```

```python
# шифрование 128 бит

def shfr(T):
    CT=T
    for i in range(0,9):
        CT=x_oper(round_key[i],CT)
        CT=s_zamn(CT)
        CT=l_oper(CT)
        print('it-',(i+1),' ',to_bytes(CT))
    CT=x_oper(round_key[9],CT)
    print('it- 10 ',to_bytes(CT))
    return CT

def rshfr(T):
    CT=T
    CT=x_oper(CT,round_key[9])
    print('it- 10 ',to_bytes(CT))
    for i in range(8,-1,-1):
        CT=obr_l(CT)
        CT=obr_zamn(CT)
        CT=x_oper(round_key[i],CT)
        print('it-',(i+1),' ',to_bytes(CT))
    return CT
```

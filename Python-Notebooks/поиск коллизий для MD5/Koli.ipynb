Задание: Разработать программу для нахождения коллизии второго рода для хэшей текстов с использованием хэш-функции, путем добавления невидимого символа

```python
### Программа:

import hashlib

def simple_hash(text):
    """
    Простая хэш-функция, возвращающая 16-битный результат.
    Используем hashlib и берём только первые 2 байта (16 бит).
    """
    return int(hashlib.md5(text.encode()).hexdigest()[:4], 16)

def find_collision(text1, text2):

    invisible_char = '\u200b'  # Невидимый символ (Zero-Width Space)
    i=0
    while True:
        # Вычисляем хэши для текущих текстов
        hash1 = simple_hash(text1)
        hash2 = simple_hash(text2)
        i=i+1

        # Проверяем, есть ли коллизия
        if hash1 == hash2:
            print("Коллизия найдена!")
            print(f"Хэш: {hash1:04x}")
            print('При ',i,' иттерации')
            break

        # Добавляем один невидимый символ в оба текста
        text1 += invisible_char
        text2 += invisible_char

# Вводим два текста
text1 = input("Введите текст 1: ")
text2 = input("Введите текст 2: ")

# Ищем коллизию
find_collision(text1, text2)
```
